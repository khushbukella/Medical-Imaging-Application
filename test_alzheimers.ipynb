```

import numpy as np
from keras.preprocessing.image import ImageDataGenerator
from keras.models import load_model
from keras.applications import VGG16

from google.colab import drive
drive.mount('/content/gdrive', force_remount=True)
```

```
datagen = ImageDataGenerator(rescale=1./255)

model = VGG16(include_top=False,weights='imagenet')

generator = datagen.flow_from_directory(
        'gdrive/My Drive/Colab Notebooks/Alzheimer_classification/Khushii/predicton', # change path here
        target_size=(150, 150),
        batch_size=1,
        class_mode=None,
        shuffle=False)
  
test_features = model.predict_generator(generator,6// 1)

np.save(open('test_features.npy', 'wb'), test_features)
```

```
model = load_model('gdrive/My Drive/Colab Notebooks/Alzheimer_classification/model.h5') # change path here
model.load_weights('gdrive/My Drive/Colab Notebooks/Alzheimer_classification/weights.h5') # change path here
```
